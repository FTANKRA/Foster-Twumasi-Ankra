## Question 2

###  The density of crimes and crime rates aren't in the data and have to be computed. Formula for computing crime density is (Crime reported/Area) and that for Crime Rate is (Crime reported/Population)* 100,000

### Merging Province Grouped Data with Population Data

```python
Province_And_Population=Province_Grouped.merge(Population_Data, on='Province')
Province_And_Population
```

```python
Province_And_Population["Density of Crime"]=Province_And_Population.Overall/Province_And_Population.Area
Province_And_Population["Crime Rate"]=(Province_And_Population.Overall/Province_And_Population.Population)*100000
Province_And_Population
```

## Contrasting Volume vs Density of Crimes.
 
 ### Due to the figure difference for volume and Density of Crime, both were not plotted on the same graph

```python
Province_And_Population[["Overall"]].plot()
plt.show()
```

```python
Province_And_Population[["Density of Crime"]].plot()
plt.show()
```

### The plot above shows the the volume of a crime affects its density. A correlation was calculated to confirm this.

```python
np.corrcoef(Province_And_Population["Overall"],Province_And_Population["Density of Crime"])
```

### Calculating Correlation Coefficient for Total Province Crime Rate and Population Density

## Comparing population density and crime rate  

```python
np.corrcoef(Province_And_Population.Density,Province_And_Population["Crime Rate"])
```

The correlation coefficient between Population Density and Crime Rate is 0.22. This means there is a fairly weak positive relationship between the two variables. Hence, Provinces with a higher Population Density have fairly high crime rates and vice versa.

### The correlation matrix does not show all coefficient, because of this, a pictorial view to the matrix was plotted to give a better view

```python
f=plt.figure(figsize=(19,15))
plt.matshow(Transpose.corr(), fignum=f.number)
plt.xticks(range(Transpose.shape[1]),Transpose.columns,fontsize=14, rotation=90)
plt.yticks(range(Transpose.shape[1]),Transpose.columns,fontsize=14)
cb=plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Crime Correlation Matrix', fontsize=16)
```

## Creating overlayed time series plots showing trend of all crime types for Kwazulu/Natal.

### Subsetting data to include only data for Kwazulu/Natal 

```python
Kwazulu=SA_Crime_Data[(SA_Crime_Data.Province == "Kwazulu/Natal")]
```

### Grouping data by category and transposing to have years as rows and crime category as columns 

```python
Kwazuluzu_Natal=Kwazulu.iloc[:,2:]
Kwazulu_grouped_transpose=Kwazuluzu_Natal.set_index('Category').groupby(['Category']).sum().transpose()
```

### Time series  plots of all crimes for Kwazulu/Natal

```python
sns.set_palette('colorblind', 8)
Kwazulu_grouped_transpose.plot(figsize=(20,10), linewidth=2, fontsize=10)
plt.xlabel('Year', fontsize=20)
```
